```python
x = np.linspace(0,2,100) # List with 100 elements varrying between 0-2

fig,ax = plt.subplots()
ax.plot(x, x, label="linear")
ax.plot(x,x**2,label="Quadratic")
ax.plot(x,x**3,label= "Cubic")
ax.set_xlabel('X label')
ax.set_ylabel('Y label')
ax.set_title("Simple Plot")
ax.legend()

x = np.linspace(2,4,100) 

plt.plot(x,x,label="linear")
plt.plot(x,x**2,label="Quadratic")
plt.plot(x,x**3,label="Cubic")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("SIMPLE PLOT")
plt.legend()
```

```python
def my_plotter(ax, data1, data2, param_dict):
    """
    A helper function to make a graph

    Parameters
    ----------
    ax : Axes
        The axes to draw to

    data1 : array
       The x data

    data2 : array
       The y data

    param_dict : dict
       Dictionary of kwargs to pass to ax.plot

    Returns
    -------
    out : list
        list of artists added
    """
    out = ax.plot(data1, data2, **param_dict)
    return out
# "**" used for unpacking dict with requried key
# "*" used for unpacking list or tuple

```

```python
data1, data2, data3, data4 = np.random.randn(4, 100)
fig, ax = plt.subplots(1, 1)
mark = {'marker': '*'}
my_plotter(ax, data1, data2, mark)
```

```python
fig, (ax1,ax2) = plt.subplots(1,2) # Create plot at row_1 with column_2
my_plotter(ax1,data1,data2,mark)
my_plotter(ax2,data3,data4, {"marker":"o"})
```

# Pyplot Tutorial

```python
plt.plot([1,2,3,4],[1,4,9,16],'r*') # args = x-data,y-data, "color Symbols"
plt.axis([0,6,0,20]) # starting and ending of X-Axis & Y-Axis
plt.show()
```

```python
t = np.arange(0.,5.,0.2)

plt.plot(t,t,"r--", t,t**2,"bs", t,t**3,'g^')
plt.show()
```

```python
data = {"a":np.arange(50),
       "c": np.random.randint(0,50,50),
       "d": np.random.randn(50)
       }
data['b'] = data["a"] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

plt.scatter("a","b",c='c',s='d',data = data)
plt.xlabel("Entry A")
plt.ylabel("Entry B")
plt.show()
```
