It is Used where the following condition are:
 1. Optimal Substrucutre:-
     Where nth answer depends on (n-1)th Answer like recursion
 2. Overlapping SubProblems (MOST IMPORTANT FOR DP):-
    Suppose we have calculate x+2 = y
    But again in future we have again calculate x+2 = y
    this kind of problem is called OVERLAPPING SUBPROBLEM

--To overcome this we store the value of certain values and this storing process is called "MEMOIZATION"

STEPS FOR DP
1. Solve using recursive relation
2. Memoization
3. Use Iteration for converting recurive

```python
# Recursive Version of DP
def fibb(n,dp):
    "O(2^n) -> O(n+1) or O(n)"
    if n == 1 or n == 0:
        return n

    if dp[n-1] == -1:
        ans1 = fibb(n-1,dp)
        dp[n-1] = ans1
    else:
        ans1 = dp[n-1]
    
    if dp[n-2] == -1:
        ans2 = fibb(n-2,dp)
        dp[n-2] = ans2
    else:
        ans2 = dp[n-2]

    myAns =  ans1+ans2
    return myAns


n = int(input())
dp = [-1 for i in range(n+1)]
print(fibb(n,dp))
```

```python
# n -> 1
def countMinStepsToOne(n,dp) :
    if n <= 1:
        return 0
    
    ans2,ans3 = 999,999
    
    if dp[n-1] == -1:
        ans1 = countMinStepsToOne(n-1,dp)
        dp[n-1] = ans1
    else:
        ans3 = dp[n-1]
    
    if n%2 == 0:
        if dp[n//2] == -1:
            ans2 = countMinStepsToOne(n//2,dp)
            dp[n//2] = ans2
        else:
            ans2 = dp[n//2]

    if n%3 == 0:
        if dp[n//3] == -1:
            ans3 = countMinStepsToOne(n//3,dp)
            dp[n//3] = ans3
        else:
            ans3 = dp[n//3]
    
    return 1+min(ans1,ans2,ans3)

def countMinStepsToOneI(n):
    if n == 0 or n == 1:
        return 0
    if n == 2 or n == 3:
        return 1
    dp = [-1 for i in range(n+1)]
    dp[0], dp[1] = 0,0
    # dp[2],dp[3] = 1,1
    i = 2
    while i<len(dp):
        ans = dp[i-1]
        if i%2 == 0:
            ans = min(ans,dp[i//2])
        if i%3 == 0:
            ans = min(ans,dp[n//3])
        dp[i] = ans+1
        i += 1

    return dp[n]
```

```python
import math,sys
def minSquares(n,dp):
    if n == 0:
        return 0

    root = int(math.sqrt(n))
    ans = sys.maxsize

    for i in range(1,root+1):
        if dp[n-(i**2)] == -1:
            currentAns = 1 + minSquares(n-(i**2),dp) 
            dp[n-(i**2)] = currentAns - 1
            
        else:
            currentAns = 1 + dp[n-(i**2)] # Saved Data Usage
        
        ans = min(ans,currentAns)
    return ans
    
    
def minSquaresI(n):
    dp = [-1 for i in range(n+1)]
    dp[0]  = 0

    for i in range(1,n+1):
        ans = sys.maxsize    
        root = int(math.sqrt(i))

        for j in range(1,root+1):
            currentAns = 1 + dp[i-(j**2)] # We can conclude this by assuming that we know it

            ans = min(ans,currentAns)
        dp[i] = ans
    return dp[n]




n = int(input())

for i in range(n+1):
    dp = [-1 for _ in range(n+1)]
    print(i,"=",minSquares(i,dp))
```

```python
# Longest Increasing Sequence (LIS)
def Lis(li):
    n = len(li)
    if n == 1:
        # Including, Overall
        return 1,1

    val = li[0]
    including_max = 0
    for i in range(1,n):
        if li[i] > val:
            further_including_max = Lis(li[i:])[0]
            including_max = max(further_including_max + 1,including_max)
    
    excluding_max = Lis(li[1:])[1]

    overall_max = max(including_max,excluding_max)

    return including_max, overall_max



def LisIndex(li,start,end):
    if start == end:
        return 1,1

    including_max = 1
    for i in range(start+1,end):
        if li[i] > li[start]:
            further_including_max = Lis(li,i,end)[0]
            including_max = max(further_including_max + 1,including_max)
    
    excluding_max = Lis(li,start+1,end)[1]

    overall_max = max(including_max,excluding_max)

    return including_max, overall_max




def LisIndexMemo(li,start,end,dp):
    
    if start == end:
        dp[start] = 1
        return 1,1

    including_max = 1
    for i in range(start+1,end):
        if li[i] > li[start]:
            if dp[i] == -1:
                ans = LisIndexMemo(li,i,end,dp)
                dp[i] = ans
                further_including_max = ans[0]
            else:
                further_including_max = dp[i][0]
            including_max = max(further_including_max + 1,including_max)
    
    if dp[start+1] == -1:
        ans = LisIndexMemo(li,start+1,end,dp)
        dp[start+1] = ans
        excluding_max = ans[1]
    else:
        excluding_max = dp[start+1][1]

    overall_max = max(including_max,excluding_max)

    return including_max, overall_max





def LisIterative(li,n):
    dp = [[0 for j in range(2)] for i in range(n+1)]
    # dp.append([1,1])
    # dp.append([0,0])

    for i in range(n-1,-1,-1):

        including_max = 1
        for j in range(i+1,n):
            if li[j] > li[i]:
                including_max = max(including_max,1+dp[j][0])
            
        dp[i][0] = including_max
        
        excluding_max = dp[i+1][1]

        overallMax = max(including_max,excluding_max)
        dp[i][1] = overallMax
    
    return dp[0][1]
    


# li = [6,3,1,2,0,7,9] # 4
# li = [1,7,2,3,4,5,6,8,9] # 8
# li = [5,4,11,1,16,8] # 3
li = [1,2,2] #2
dp = [-1 for i in range(len(li)+1)]
print(LisIterative(li,len(li)))
```
