# Sorting DataFrames

```python
import pandas as pd
import numpy as np
```

```python
data_BM = pd.read_csv("Pandas_Part_1/bigmart_data.csv")

#Drop the null values
data_BM = data_BM.dropna(how="any")

data_BM.head()
```

```python
# pd.sort_values = To sort pandas DF by one or more columns
sorted_data = data_BM.sort_values(by="Outlet_Establishment_Year") # Ascending Columns
sorted_data[-5:]
```

```python
data_BM.sort_values(by="Outlet_Establishment_Year",ascending=False,inplace=True) #Inplace means not to return anything but to make changes in real DF
data_BM[-5:]
```
